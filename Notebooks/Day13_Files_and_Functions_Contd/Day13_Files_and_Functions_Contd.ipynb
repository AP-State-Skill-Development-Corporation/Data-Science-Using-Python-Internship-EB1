```python
def freq_words(data):
    data = data.split()
    print(len(data))
    
    uniq = set(data)
    print(len(uniq))
    
    freq = {}
    for word in uniq:
        freq[word] = data.count(word)
    return freq
```

```python
a = 55

def fun(b):
    print(a)
    print(b)
    
print(a)
fun(123)
print(b)
```

```python
a = 55

def fun(b):
    print(a) #-> globalVar
    print(b) #-> local
    def fun2(c):
        print(a) #-> globalVar
        print(b) #-> LocalVar
        print(c)
    fun2(365)
    print(c) #-> nonLocal
    
print(a)
fun(123)
print(b)
```

```python
def call_value(val):
    return val
```

```python
a = 55
call_value(a)
```

```python
def call_value(val):
    print(val, id(val))
    val = 625
    return val, id(val)
```

```python
a = 55
call_value(a)
```

```python
def call_ref(val):
    print(val, id(val))
    val.append(625)
    return val, id(val)
```

```python
li = [1,2,3,4,5]
print(id(li))

print(call_ref(li))

print(li)
```

```python
li = [1,2,3,4,5]
print(id(li))

print(call_ref(li.copy())) # Call ref

print(li)
```

```python
print(id(li[:]))
```
