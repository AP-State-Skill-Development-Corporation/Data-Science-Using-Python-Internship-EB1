 ## Inheritance
    - Inheritance is defined as  the capability of one class to derive or inherit the Properties from some other class
- Properties:
    - represents real _world relationships
    - reusability of predifiend code.
    - transitive in nature

```python
class parent:
    def f1(self):
        print(" In parent class")
class child(parent):
    def f2(self):
        print("In child class")
        
```

```python
dir(child)
```

```python
obj=child()
```

```python
obj.f1()
```

```python
obj.f2()
```

## Types of inheritance
1. single  Inheritance
2. Multiple Inheritance
3. Multi-Level Inheritance
4. Hierarchical Inheritance
5. Hybrid Inheritance

## single  Inheritance

```python

class parent:
    def f1(self):
        print(" In parent class")
class child(parent):
    def f2(self):
        print("In child class")
x=child() 
x.f2()
x.f1()
```

## Multiple Inheritance

- one derived or child classInherit Properties  from one or more super classes

```python
class A:
    def f1(self):
        print("class A function f1")
class B:
    def f2(self):
        print("class B  function f2")
class C(A,B):
    def f3(self):
        print("Class C function f3")
obj=C()
    
```

```python
dir(obj)
```

```python
obj.f2()
```

```python
obj.f1()
```

```python
obj.f3()
```

## Multi_Level Inheritance

- features of the base class and derived class are future inherited into the new derived class.

```python
class grandparent:
    def f1(self):
        print("class A function f1")
class parent(grandparent):
    def f2(self):
        print("class B  function f2")
class Child(parent):
    def f3(self):
        print("Class C function f3")
y=Child()
```

```python
y.f1()
```

```python
y.f2()
```

```python
y.f3()
```

## Hierachical Inheritance
- when more than one dervied  classes Inherit properties from single base(parent)

```python
class p1:
    def fun1(self):
        print("In  parent class ")
class c1(p1):
    def fun2(self):
        print("In  child 1 class ")
class c2(p1):
    def fun3(self):
        print("In  child 2 class ")  
z1=c1()
z2=c2()
```

```python
z1.fun1()
```

```python
z1.fun2()
```

```python
z2.fun1()
```

```python
z2.fun3()
```

## Hybrid Inheritance
- Inheritance consisting of multiple types of Inheritance 

```python
class clg:
    def f1(self):
        print("function in colleage class")
class student1(clg):
    def f2(self):
        print("function in student 1 class")
class student2(student1,clg):
    def f3(self):
        print("function in student 2 class")
s=student2()
               
    
```

```python
s.f1()
```

```python
s.f2()
```

## Data Encapsulation

- Encapsulation is one of the fundamental concept in OOPs
- Describes the data wrapping with one unit
- Restrict  accessing of varaibles and methods of class directlty.
 - Private variables
 - Private Methods
     - use '__' to difine a private methods or private variables

## Private Variables

```python

class python:
    def __init__(self):
        self.__a='apssdc'
    def view(self):
        print(self.__a)
    def v1(self):
        self.__a=1000
        print(self.__a)

x=python()        
```

```python
x.view()
```

```python
x.v1()
```

```python
x.__a
```

```python
class m1:
    __a=10
    def add(self,b):
        sum=self.__a+b
        return (sum)
ob=m1()
```

```python
ob.add(121)
```

```python
ob.__a
```

## Private  Methods 


```python
class car:
    def __init__(self,x):
        self.x=x
        self.__update(x)
    def drive (self):
        print("driving")
    def __update(self,x):
        print(self.x)
        print("update your software")
c=car(12)
```

```python
c.drive()
```

```python
c.__update(2)
```

## Constructor in Inheritence


```python
class p1:
    def __init__(self):
        print("p1 constructor")
    def fun1(self):
        print("In  parent class ")
class c1(p1):
    def __init__(self):
        super().__init__()
        print("belongs to c1 constructor")
    def fun2(self):
        print("In  child 1 class ")
a1=c1()
```

```python
class p1:
    def __init__(self):
        print("p1 constructor")
    def fun1(self):
        print("In  parent class ")
class p2:
    def __init__(self):
        print("p2 constructor")
    def fun2(self):
        print("In  parent2 class ")       
class c1(p1,p2):
    def __init__(self):
        super().__init__()
        print("belongs to c1 constructor")
    def fun3(self):
        super().fun2()
        print("In  child 1 class ")
a1=c1()
```

```python
a1.fun3()
```
